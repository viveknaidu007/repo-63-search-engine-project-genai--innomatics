```python
from sentence_transformers import SentenceTransformer
```

```python
# Sample text
sample_text = "This is a sample sentence to convert into vectors."

# Initialize a BERT-based SentenceTransformer model
model = SentenceTransformer('bert-base-nli-mean-tokens')

# Convert the sample text into vectors
vector = model.encode(sample_text)

# Print the vector
print(vector)
```

```python
from transformers import BertTokenizer, BertModel
import torch

# Load pre-trained BERT model and tokenizer
model_name = 'bert-base-uncased'  # You can specify different BERT variants
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# Example sentence
sentence = "Embeddings are cool!"

# Tokenize input sentence
tokens = tokenizer.tokenize(sentence)
token_ids = tokenizer.convert_tokens_to_ids(tokens)

# Convert tokens to tensor
input_ids = torch.tensor([token_ids])

# Forward pass through BERT model
with torch.no_grad():
    outputs = model(input_ids)

# Get the hidden states from the output
hidden_states = outputs[0]

# Print embeddings for each token
for i, token in enumerate(tokens):
    print(f"Token: {token}")
    print(f"Embedding: {hidden_states[0, i].tolist()}")
    print()
```
